```python
import kplr
import numpy as np
import matplotlib.pyplot as pl
from simplexy import simplexy
```

```python
client = kplr.API()
tpf = client.k2_star(202137899).get_target_pixel_files()[0]
```

```python
data = tpf.read()
time = data["TIME"]
flux = data["FLUX"]

t = time[~np.isnan(time)]
i = np.argmax(np.diff(t))
i = np.arange(len(time))[~np.isnan(time)][i+1]
print(i)
time = time[i:]
flux = flux[i:]
```

```python
def run_one(img):
    m = np.isfinite(img)
    m[m] *= (img[m] > 0.0)
    if not np.any(m):
        return None
    img[~m] = np.median(img[m])
    return np.array([list(row) for row in simplexy(img)])
```

```python
c = run_one(flux[0])
coords = np.nan + np.zeros((len(flux), len(c), 4))
coords[0, :, :] = np.array(c)

for i in range(1, len(flux)):
    img = flux[i]
    c = run_one(img)
    if c is None:
        coords[i] = coords[i-1]
        continue
    r2 = (c[:, 0][:, None] - coords[i-1, :, 0])**2 + (c[:, 1][:, None] - coords[i-1, :, 1])**2
    r2[~np.isfinite(r2)] = np.inf
    ix = np.arange(len(c))
    iy = np.argmin(r2, axis=1)
    r2 = r2[(ix, iy)]
    print(i, iy)
    m = r2 < 4.0
    coords[i, iy[m], :] = c[ix[m]]
    # print(coords[i])
    
print(coords[:, 0, 0])
pl.plot(time, coords[:, 0, 2], ".k")
```
